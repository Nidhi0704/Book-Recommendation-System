## Popularity Based Recommender System:-

```python
rating_with_name = Ratings.merge(Books, on='ISBN')
```

```python
rating_with_name.head()
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.head()
```

```python
num_rating_df.rename(columns={'Book-Rating': 'num_rating'}, inplace=True)
```

```python
num_rating_df
```

```python
rating_with_name['Book-Rating'] = pd.to_numeric(rating_with_name['Book-Rating'], errors='coerce')
mean_ratings = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
mean_ratings.rename(columns={'Book-Rating':'mean_rating'}, inplace=True)
```

```python
mean_ratings.head()
```

```python
popular_df = num_rating_df.merge(mean_ratings, on='Book-Title')
```

```python
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_rating'] >= 250].sort_values('mean_rating', ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(Books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','Image-URL-M','num_rating','mean_rating']]
```

```python
popular_df.head()
```

## Collaborative Filtering Based Recommedation System

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating']>200
```

```python
rating_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(rating_users)]
```

```python
y  = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_book = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating
```

```python
final_rating.duplicated().sum()
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar_score = cosine_similarity(pt)
```

```python
similar_score.shape
```

```python
def recommend(book_name):
    # Convert both book_name and pt.index to lowercase to avoid case sensitivity issues
    book_name = book_name.strip().lower()
    pt.index = pt.index.str.strip().str.lower()

    # Clean book titles in Books dataframe and drop duplicates
    Books['Book-Title'] = Books['Book-Title'].str.strip().str.lower()
    Books.drop_duplicates(subset='Book-Title', inplace=True)

    # Check if the book exists in the index
    if book_name not in pt.index:
        print(f"'{book_name}' not found in the book list.")
        return

    ## index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similar_score[index])), key=lambda x: x[1], reverse=True)[1:8]

    data = []
    for i in similar_items:
        item = []
        temp_df = Books[Books['Book-Title'] == pt.index[i[0]]]

        # Debugging output to check for duplicates
        print(temp_df)

        if not temp_df.empty:
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Author')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Image-URL-M')['Image-URL-M'].values))
            data.append(item)

    return data
```

```python
recommend('1984')
```

```python
recommend('Message in a bottle')
```

```python
recommend('The Notebook')
```
